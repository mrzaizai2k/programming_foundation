## Transform text columns

```python
df.shape
```

```python
df.sample(2)
```

```python
corpus = df.text.tolist()
len(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
counter1 = CountVectorizer()
counter2 = CountVectorizer(max_features=1000)
counter1.fit(corpus)
counter2.fit(corpus)

vectorizer1 = TfidfVectorizer()
vectorizer2 = TfidfVectorizer(max_features=1000)
vectorizer1.fit(corpus)
vectorizer2.fit(corpus)
```

```python
X1 = counter1.transform(corpus)
X2 = counter2.transform(corpus)
Y1 = vectorizer1.transform(corpus)
Y2 = vectorizer2.transform(corpus)
```

```python
X1np = X1.toarray()
X2np = X2.toarray()
X1np.shape, X2np.shape
```

```python
Y1np = Y1.toarray()
Y2np = Y2.toarray()
Y1np.shape, Y2np.shape
```

```python
X2np[:4, :4]
```

```python
Y2np[:4, :4]
```

```python
for idx, (k, v) in enumerate(counter2.vocabulary_.items()):
  print(f"{k:20s}: {v}")
  if idx > 10:
    break
```

```python
len(counter2.vocabulary_.items())
```

```python
X2np.shape
```

```python
X2np.max(axis=-1, keepdims=True).shape
TF = X2np/X2np.sum(axis=-1, keepdims=True)
TF.shape
```

```python
IDF_denom = X2np.sum(axis=0, keepdims=True)
IDF_denom.shape
IDF = np.log(len(X2np)/IDF_denom)
TFIDF = TF*IDF
TFIDF.shape
```
